```python
import pandas as pd

# clase de la empreza dedicada a proyectos de construccion
class Empreza:
    def __init__(self, nombre, cuenta_bancaria):
        self.nombre = nombre
        self.cuenta_bancaria = cuenta_bancaria

class Proyecto(Empreza):
    def __init__(self, nombre, cuenta_bancaria, nombre_proyecto, mision_proyecto):
        super().__init__(nombre, cuenta_bancaria)
        self.nombre_proyecto = nombre_proyecto
        self.mision_proyecto = mision_proyecto
    # va actualizando el saldo de la cuenta dependido de las salidas y entradas
    def cuenta_empresa(self, saldo_actual, entrada_salida, tipo):
        if tipo == 'entrada':
            cuenta = saldo_actual + entrada_salida
            return cuenta
        elif tipo == 'salida':
            cuenta = saldo_actual - entrada_salida
            return cuenta
    
    # Muestra la tabla de los empleado que hay en la empreza
    def mostrar_datos_proyecto(self, registro_empleados, saldo_actual):
        data = []
        for nombre, datos in registro_empleados.items():
            labor = datos.get("labor", "N/A")
            horas_extras = datos.get("horas_extras", 0)
            salario = datos.get("salario", 0)
            bonificacion = datos.get("bonificacion", 0)
            salario_final = datos.get("salario_final", 0)
            data.append([nombre, datos["tipo empleado"], datos["año de nacimiento"], labor, horas_extras, salario, bonificacion, salario_final])
        df = pd.DataFrame(data, columns=["Nombre", "Tipo Empleado", "Año de Nacimiento", "Labor", "Horas Extras", "Salario", "Bonificación", "Salario Final"])
        df.loc[len(df.index)] = ["Saldo Actual", "", "", "", "", "", "", saldo_actual]
        print(df)
    
    # actualiza la cuenta bancaria
    def actualizar_cuenta(self, total_salida):
        self.cuenta_bancaria -= total_salida
        return self.cuenta_bancaria

class Empleados:
    # registra a las paersonas que estan trabajndo en obra
    def registrar_persona(self, cantidad):
        registro_empleados = {}
        registro_contratistas = {}
        for i in range(cantidad):
            nombre = input(f"ingrese el nombre de la persona {i+1}: ")
            año_de_nacimiento = input("ingrese su fecha de nacimiento: ")
            tipo_empleado = input("ingrese que tipo de empleado es: ")
            if tipo_empleado == "contratista":
                registro_contratistas[nombre] = {"tipo empleado": tipo_empleado, "año de nacimiento": año_de_nacimiento}
            elif tipo_empleado == "empleado":
                registro_empleados[nombre] = {"tipo empleado": tipo_empleado, "año de nacimiento": año_de_nacimiento}
        return registro_contratistas, registro_empleados

class EmpRegulares:
    def __init__(self, labor, horas_extras, domingos_trabajados):
        self.labor = labor
        self.horas_extras = horas_extras
        self.domingos_trabajados = domingos_trabajados
    
    # calcula el salario dependiendo de su puesto laboral
    def calcular_salario(self):
        if self.labor == "auxiliar":
            salario = 1400000
        elif self.labor == "oficial":
            salario = 2200000
        elif self.labor == "maestro":
            salario = 3500000
        return salario
    
    # depende de las horas extras puede recibir mas dinero
    def bonificacion(self):
        horas_extras_domingos = self.domingos_trabajados * 6
        total_horas_extras = self.horas_extras + horas_extras_domingos
        valor_hora_extra = (self.calcular_salario() / 160) * 1.5
        bonificacion = total_horas_extras * valor_hora_extra
        return bonificacion

    # calcula el salario final del empleado
    def salario_final(self):
        salario = self.calcular_salario()
        bonificacion = self.bonificacion()
        salario_final = salario + bonificacion
        return salario_final

class Contratistas:
    def __init__(self, casas_construidas):
        self.casas_construidas = casas_construidas 
        
    # por cada casa vendida reciben cierta cantidad
    def pago_por_casa(self):
        pago_por_casa = 50000000  
        total_pago = self.casas_construidas * pago_por_casa
        return total_pago

def main():
    nomina = {}
    print("EMPREZA DE CONTRUCCIONES")
    nombre_empreza = input(" -Ingrese el nombre de la empreza: ")
    saldo_actual = int(input(" -Ingrese el saldo actual de la cuenta: "))
    nombre_proyecto = input("\t-Ingrese el nombre del proyecto actual: ")
    mision_proyecto = input("\t-Ingrese su mision principal: ")
    
    empreza = Proyecto(nombre_empreza, saldo_actual, nombre_proyecto, mision_proyecto)
    
    print(f"\n----------------BIENVENIDO A {empreza.nombre}-----------------\n")
    print(f"\t\tSaldo cuenta: {empreza.cuenta_bancaria:,}\n")
    print(f"\t-Actualmente trabajamos en el proyecto {empreza.nombre_proyecto}")
    print(f"\n-su mision principal es: {empreza.mision_proyecto}")
    
    empleado_contratista = Empleados()
    cantidad_empleado_contratista = int(input("Cuantos empleados y contratistas ingresaron a la empreza: "))
    registro_contratista, registro_empleado = empleado_contratista.registrar_persona(cantidad_empleado_contratista)
        
    print("LABORES EN LA CONTRUCCION Y SU SALARIO")
    puestos_laborales = ["auxiliar", "oficial", "maestro"]
    
    total_salida = 0
        
    for nombre, dato in registro_empleado.items():
        print(f"Ingrese la labor de {nombre}: ")
        for i, puesto in enumerate(puestos_laborales, 1):
            print(f"{i} - {puesto} de obra")
        labor = int(input())
        horas_extras = int(input("Cuantas horas extras trabajas al mes: "))
        domingos_trabajados = int(input("Cuantos domingos ha trabajado: "))
            
        emp_regulares = EmpRegulares(puestos_laborales[labor - 1], horas_extras, domingos_trabajados)
        salario_final = emp_regulares.salario_final()
            
        dato["salario"] = emp_regulares.calcular_salario()
        dato["labor"] = emp_regulares.labor
        dato["horas_extras"] = emp_regulares.horas_extras
        dato["bonificacion"] = emp_regulares.bonificacion()
        dato["salario_final"] = salario_final
        
        total_salida += salario_final
    
    for nombre, dato in registro_contratista.items():
        casas_construidas = int(input(f"Cuantas casas fueron construidas por {nombre}: "))
        contratista = Contratistas(casas_construidas)
        salario_contratista = contratista.pago_por_casa()
        
        dato["salario"] = salario_contratista
        dato["salario_final"] = salario_contratista
        
        total_salida += salario_contratista
    
    print(" ")
    empreza.mostrar_datos_proyecto({**registro_empleado, **registro_contratista}, empreza.cuenta_bancaria)
    nueva_cuenta = empreza.actualizar_cuenta(total_salida)
    print(f"La cuenta bancaria actualizada es: {nueva_cuenta:,}")

if __name__ == "__main__":
    main()
```
